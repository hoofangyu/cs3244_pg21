```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the ROC data from the CSV file
df = pd.read_csv('roc_curve_data.csv')

# Display basic information about the dataframe
print(df.head())  # Check the first few rows
print(df.info())  # Get info about columns and data types

# Check for any null or missing values
print(df.isnull().sum())


# Replace 'inf' with a large number (e.g., 9999)
df['Threshold'].replace('inf', 9999, inplace=True)
df['Threshold'] = df['Threshold'].astype(float)


# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(df['FPR'], df['TPR'])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.grid()
plt.show()
```
