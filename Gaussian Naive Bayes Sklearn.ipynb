```python
##This code is completed with the aid of ChatGPT

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
```

```python
train = pd.read_pickle("train.pkl").dropna()
test = pd.read_pickle("test.pkl").dropna()

train.head()
```

```python
comment_train = train['comment']
comment_test = test['comment']
label_train = train['label']
label_test = test['label']
#comment_train, comment_test, label_train, label_test = train_test_split(comment, label, test_size = 0.2, random_state = 99)
```

```python
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(comment_train)
X_test = vectorizer.transform(comment_test)
```

```python
classifier = MultinomialNB()
classifier.fit(X_train, label_train)

predictions = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(label_test, predictions)
report = classification_report(label_test, predictions)
```
